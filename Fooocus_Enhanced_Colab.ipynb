## Cell 4️⃣: Setup Colab-specific Launcher

```python
# Create Colab-specific launcher with ngrok
colab_launcher = '''
#!/usr/bin/env python3
"""
Fooocus Enhanced - Colab Launcher
Optimized for Google Colab with ngrok tunneling
"""

import os
import sys
import subprocess
from pathlib import Path

# Setup paths
root = os.path.dirname(os.path.abspath(__file__))
sys.path.insert(0, root)
os.chdir(root)

# Configure for Colab
os.environ["PYTORCH_ENABLE_MPS_FALLBACK"] = "1"
os.environ["PYTORCH_MPS_HIGH_WATERMARK_RATIO"] = "0.0"
os.environ["GRADIO_SERVER_PORT"] = "7865"

print("[Fooocus] 🚀 Launching Enhanced WebUI on Colab...")
print(f"[Fooocus] Python: {sys.version}")
print(f"[Fooocus] Working directory: {os.getcwd()}")

try:
    # Import Gradio and webui_demo
    print("[Fooocus] Importing modules...")
    import gradio as gr
    from webui_demo import create_demo_interface
    
    print("[Fooocus] Creating demo interface...")
    demo = create_demo_interface()
    
    # Setup ngrok tunnel
    print("[Fooocus] Setting up ngrok tunnel...")
    import pyngrok
    from pyngrok import ngrok
    
    # Set ngrok auth token (optional for more tunnels)
    # ngrok.set_auth_token("YOUR_TOKEN_HERE")
    
    # Create tunnel
    tunnel = ngrok.connect(7865, "http")
    print(f"\n[Fooocus] ✅ Tunnel created!")
    print(f"[Fooocus] 🔗 Public URL: {tunnel.public_url}")
    print(f"[Fooocus] 📌 Keep this tab open to maintain the tunnel\\n")
    
    # Launch Gradio
    print("[Fooocus] Starting Gradio server...")
    demo.launch(
        server_name="0.0.0.0",  # Listen on all interfaces for Colab
        server_port=7865,
        ssl_keyfile=None,
        ssl_certfile=None,
        show_error=True,
        share=False,  # We're using ngrok instead
        theme=gr.themes.Soft()
    )
    
except Exception as e:
    print(f"[ERROR] Failed to launch: {type(e).__name__}: {e}")
    import traceback
    traceback.print_exc()
'''

# Write the launcher
with open("launch_colab.py", "w") as f:
    f.write(colab_launcher)

print("✅ Created launch_colab.py")
```
